# Analysis- Kerala

Kerala is a state on the southwestern Malabar Coast of India. It was formed on 1 November 1956, following the passage of the States Reorganisation Act, by combining Malayalam-speaking regions of the erstwhile states of Travancore-Cochin and Madras.

Kerala is considered to be a state with the highest literacy rates. It is India's number one state in health facilities. 
These ascpects might not sound important, but play an indirect role in their fight against the Novel Corona virus (Covid-19).

```python
kerala_case_counts = pd.read_csv('complete.csv')[pd.read_csv('complete.csv')['Name of State / UT'] == 'Kerala']
```

```python
# kerala_case_counts['Date'] = pd.to_datetime(kerala_case_counts['Date'])
kerala_case_counts.tail()
```

# Total cases reported, deaths and cured/migrated/discharged seen in Kerala from 30.01.2020 to 07.05.2020

1. Total cases                - 503
2. Deaths                     - 4
3. Cured/Migrated/Discharged  - 469

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,8))
sns.lineplot(x='Date', y='Total Confirmed cases', data=kerala_case_counts, label='Total Confirmed cases')
sns.lineplot(x='Date', y='Cured/Discharged/Migrated', data=kerala_case_counts, label='Cured/Discharged/Migrated')
sns.lineplot(x='Date', y='Death', data=kerala_case_counts, label='Death')
plt.legend()
```

```python
l = []
l.append(kerala_case_counts['Total Confirmed cases'].iloc[0])
for i in range(1, 99):
    l.append(kerala_case_counts['Total Confirmed cases'].iloc[i] - kerala_case_counts['Total Confirmed cases'].iloc[i-1])
kerala_case_counts['Daily Reported Cases'] = l

l = []
l.append(kerala_case_counts['Death'].iloc[0])
for i in range(1, 99):
    l.append(kerala_case_counts['Death'].iloc[i] - kerala_case_counts['Death'].iloc[i-1])
kerala_case_counts['Daily Reported deaths'] = l

l = []
l.append(kerala_case_counts['Cured/Discharged/Migrated'].iloc[0])
for i in range(1, 99):
    l.append(kerala_case_counts['Cured/Discharged/Migrated'].iloc[i] - kerala_case_counts['Cured/Discharged/Migrated'].iloc[i-1])
kerala_case_counts['Daily Reported CDM'] = l
```

# Daily cases, deaths and cures/migrations/discharges seen in Kerala

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,8))
sns.lineplot(x='Date', y='Daily Reported Cases', data=kerala_case_counts, label='Daily Reported Cases')
sns.lineplot(x='Date', y='Daily Reported CDM', data=kerala_case_counts, label='Daily Reported CDM')
sns.lineplot(x='Date', y='Daily Reported deaths', data=kerala_case_counts, label='Daily Reported deaths')
plt.legend()
```

```python
c = (kerala_case_counts['Total Confirmed cases'].max() - (kerala_case_counts['Cured/Discharged/Migrated'].max() + kerala_case_counts['Death'].max()))/kerala_case_counts['Total Confirmed cases'].max()
r = kerala_case_counts['Cured/Discharged/Migrated'].max()/kerala_case_counts['Total Confirmed cases'].max()
d = kerala_case_counts['Death'].max()/kerala_case_counts['Total Confirmed cases'].max()
```

# Cases distributions (Positive cases, Cured/Migrated/Discharged, Deaths) 

```python
plt.figure(figsize=(7, 7))
plt.pie([c*100, r*100, d*100], autopct='%1.1f%%', labels=['Confirmed', 'Cured/Migrated/Discharged', 'Deaths'])
plt.show()
```

Most of the cases have no been migrated, discharged or cured (93.2%), the state of Kerala has almost flattened their curve. Considering Kerala being a reference model, many other states and countries can take their idea and help their states improve in this situation of pandemic.

**The Kerala's fight against Covid has been successful, but there are many aspects that helped them improve.**
1. Older experiences: Kerala before this pandemic had faced another deadly disease, the Nipah virus. Just like Corona virus, this virus too had originated from bats. This virus had a much higher mortality rate when compared to Corona virus. Almost 19 deaths were seen. This had helped Kerala fight against Corona much easier as they had prior experience on contact tracing. As the health minister of Kerala, K.K. Shailaja, in an interview with India Today & NDTV told that they had started tracing their citizens staying outside India, specially in Wuhan. 
2. Health system and facilities: Kerala is considered to have the best health facilities than any other state in India. 
3. Literacy: Kerala alone has a literacy rate of 93.21%. Better education helped create better awareness among the people of Kerala.

Kerala had started their tracing much much before any stated had thought of. They considered this problem as a real problem. They have been lucky too because of prior experience, but even if we consider that Kerala hadn't faced something like Nipah, they still would have planned things better. When Nipah hit Kerala, it was their first time for something like this. Kerala was lauded by WHO on their fight against Nipah.
